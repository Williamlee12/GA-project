```python
handles = []


for item in df.accounts:
    if 'office' in item.keys() and 'campaign' in item.keys():
        if len(item['office']) > 1 and len(item['campaign']) > 1:
            handles.append([item['index'], item['office'][0]['screen_name'], 'no'])
            handles.append([item['index'], item['office'][1]['screen_name'], 'no'])
            handles.append([item['index'], item['campaign'][0]['screen_name'], 'yes'])
            handles.append([item['index'], item['campaign'][1]['screen_name'], 'yes'])
        elif len(item['office']) > 1 and len(item['campaign']) == 1:
            handles.append([item['index'], item['office'][0]['screen_name'], 'no'])
            handles.append([item['index'], item['office'][1]['screen_name'], 'no'])
            handles.append([item['index'], item['campaign'][0]['screen_name'], 'yes'])
        elif len(item['office']) == 1 and len(item['campaign']) > 1:
            handles.append([item['index'], item['office'][0]['screen_name'], 'no'])
            handles.append([item['index'], item['campaign'][0]['screen_name'], 'yes'])
            handles.append([item['index'], item['campaign'][1]['screen_name'], 'yes'])
        elif len(item['office']) == 1 and len(item['campaign']) == 1:
            handles.append([item['index'], item['office'][0]['screen_name'], 'no'])
            handles.append([item['index'], item['campaign'][0]['screen_name'], 'yes'])
        elif len(item['office']) == 0 and len(item['campaign']) == 1:
            handles.append([item['index'], item['campaign'][0]['screen_name'], 'yes'])
        elif len(item['office']) == 1 and len(item['campaign']) == 0:
            handles.append([item['index'], item['office'][0]['screen_name'], 'no'])
    elif 'office' in item.keys():
        if len(item['office']) > 1:
            handles.append([item['index'], item['office'][0]['screen_name'], 'no'])
            handles.append([item['index'], item['office'][1]['screen_name'], 'no'])
        else:
            handles.append([item['index'], item['office'][0]['screen_name'], 'no'])
    else:
        if len(item['campaign']) > 1:
            handles.append([item['index'], item['campaign'][0]['screen_name'], 'yes'])
            handles.append([item['index'], item['campaign'][1]['screen_name'], 'yes'])
        else:
            handles.append([item['index'], item['campaign'][0]['screen_name'], 'yes'])
        
len(handles)
           
```
