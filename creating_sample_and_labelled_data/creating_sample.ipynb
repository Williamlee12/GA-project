```python
import pandas as pd
```

```python
df = pd.read_csv("../csv_creation/final_data_new.csv", encoding="utf-8")
```

```python
df.drop('Unnamed: 0', 1, inplace=True)
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.fillna('NaN', inplace=True)
```

```python
tweets = []

for x in df.text:
    if len(x) > 100:
        tweets.append(x.encode('utf-8'))
```

```python
len(tweets)
```

```python
import random
```

```python
tweets = random.sample(tweets, 3000)
```

```python
df1 = pd.DataFrame(tweets, columns = ['tweet'])
```

```python
df1
```

```python
df1.to_csv('training_data.csv', index=False, encoding='utf-8')
```
