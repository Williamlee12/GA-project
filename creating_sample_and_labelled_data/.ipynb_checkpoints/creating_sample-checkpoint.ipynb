```python
import pandas as pd
```

```python
df = pd.read_csv("../csv_creation/final_data.csv", encoding="utf-8")
```

```python
df.drop('Unnamed: 0', 1, inplace=True)
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.fillna('NaN', inplace=True)
```

```python
tweets = []

for x in df.text:
    if len(x) > 100:
        tweets.append(x.encode('utf-8'))
```

```python
len(tweets)
```

```python
import random
```

```python
tweets = random.sample(tweets, 3000)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
list_of_scores = []

for tweet in tweets:
    vs = analyzer.polarity_scores(tweet)
    list_of_scores.append([tweet, vs])
```

```python
df1 = pd.DataFrame(list_of_scores, columns = ['tweet', 'scores'])
```

```python
scores = []

for x in df1.scores:
    if x['compound'] > 0.8:
        scores.append(3)
    elif x['compound'] < -0.8:
        scores.append(1)
    elif x['compound'] > -0.2 and x['compound'] < 0.2:
        scores.append(2)
    else:
        scores.append('ambiguous')
```

```python
df1['sentiment'] = scores
```

```python
df1.sentiment.value_counts()
```

```python
df1.sort_values('sentiment', inplace=True)
```

```python
df1.reset_index(inplace=True, drop=True)
```

```python
df1
```

```python
df1.to_csv('vader_labelled_data.csv', index=False, encoding='utf-8')
```
