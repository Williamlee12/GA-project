```python
import pandas as pd
```

```python
df = pd.read_json('final_data.json')
```

```python
tweets = []

for x in df.text:
    if len(x) > 100:
        tweets.append(x.encode('utf-8'))
```

```python
len(tweets)
```

```python
import random
```

```python
tweets = random.sample(tweets, 2000)
```

```python
len(tweets)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
lists = []

for tweet in tweets:
    vs = analyzer.polarity_scores(tweet)
    lists.append([tweet, vs])
```

```python
lists
```

```python
df = pd.DataFrame(lists, columns = ['tweet', 'scores'])
```

```python
scores = []

for x in df.scores:
    if x['compound'] > 0.8:
        scores.append(0)
    elif x['compound'] < -0.8:
        scores.append(-1)
    elif x['compound'] > -0.2 and x['compound'] < 0.2:
        scores.append(0)
    else:
        scores.append('ambiguous')
```

```python
scores
```

```python
df['sentiment'] = scores
```

```python
df.head()
```

```python
df.sentiment.value_counts()
```

```python
df.to_csv('labelled_data.csv')
```
